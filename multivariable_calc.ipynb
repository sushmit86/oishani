```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python


# Define the function alpha
def alpha(t):
  return (t, t**2)

# Create a range of t values
t_values = np.linspace(0, 2, 100) # 100 points between 0 and 2

# Calculate the corresponding x and y values
x_values = [alpha(t)[0] for t in t_values]
y_values = [alpha(t)[1] for t in t_values]

# Create the plot
plt.plot(x_values, y_values)
plt.xlabel('t')
plt.ylabel('t^2')
plt.title('Plot of alpha(t) = (t, t^2)')
plt.grid(True)
plt.show()
```
